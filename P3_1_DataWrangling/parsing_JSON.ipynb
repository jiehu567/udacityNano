```python
# should do pip install requests
import json
import requests
```

```python
BASE_URL = "http://musicbrainz.org/ws/2/"
ARTIST_URL = BASE_URL + "artist/"
```

```python
query_type = {  "simple": {},
                "atr": {"inc": "aliases+tags+ratings"},
                "aliases": {"inc": "aliases"},
                "releases": {"inc": "releases"}}
```

```python
query_type
```

```python
def query_site(url, params, uid="", fmt="json"):
    params["fmt"] = fmt
    r = requests.get(url + uid, params=params)
    print "requesting", r.url

    if r.status_code == requests.codes.ok:
        return r.json()
    else:
        r.raise_for_status()
```

```python
def query_by_name(url, params, name):
    params["query"] = "artist:" + name
    return query_site(url, params)
```

```python
def pretty_print(data, indent=4):
    if type(data) == dict:
        print json.dumps(data, indent=indent, sort_keys=True)
    else:
        print data
```

```python
# Nirvana disambiguation
results = query_by_name(ARTIST_URL, query_type["simple"], "Nirvana")
pretty_print(results) 
```

```python
for item in results['artists']:
    if item.has_key(u'disambiguation') and item.has_key('country'):
        if item['country'] == 'US': 
            print item[u'disambiguation']
```

```python
results = query_by_name(ARTIST_URL, query_type["simple"], "One Direction")
pretty_print(results) 
```

```python
for item in results['artists']:
    if item['name'] == 'One Direction':
        print item['life-span']
```

```python

results = query_by_name(ARTIST_URL, query_type["simple"], "First Aid Kit")
pretty_print(results)   # a dictionary


artist_id = results["artists"][1]["id"]
print "\nARTIST:"
pretty_print(results["artists"][1])

artist_data = query_site(ARTIST_URL, query_type["releases"], artist_id)
releases = artist_data["releases"]
print "\nONE RELEASE:"
pretty_print(releases[0], indent=2)
release_titles = [r["title"] for r in releases]

print "\nALL TITLES:"
for t in release_titles:
    print t

```
